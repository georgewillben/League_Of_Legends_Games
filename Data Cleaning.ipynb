# Data Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import StratifiedShuffleSplit, train_test_split
from sklearn.impute import SimpleImputer
```

# 1.1 Load Data

```python
df = pd.read_csv("high_diamond_ranked_10min.csv")
df.head()
```

```python
df.info()
```

# 2.1 Inspect Data for Errors

```python
min_maxes = []
for col in df.columns:
    min_maxes.append({"Column":col, "Min":df[col].min(), "Max":df[col].max()})
    
pd.DataFrame(min_maxes)
```

All these values seem to not be errors. There are no negative measurements

# 3.1 Inspect Distribution Of Columns

```python
df.hist(figsize=(25, 15))
plt.tight_layout()
plt.show()
```

Many of the columns seem to be normally distributed.

# 3.2 See If Target Class Is Imbalanced

```python
blue_win = df.blueWins.sum()
red_win = sum([1 if bluewin == 0 else 0 for bluewin in df.blueWins])

plt.figure(figsize=(15, 5))
bars = plt.bar(["Blue Wins", "Red Wins"], [blue_win, red_win])
plt.title("Wins by Colour")
bars[0].set_color("blue")
bars[1].set_color("red")
plt.show()
```
